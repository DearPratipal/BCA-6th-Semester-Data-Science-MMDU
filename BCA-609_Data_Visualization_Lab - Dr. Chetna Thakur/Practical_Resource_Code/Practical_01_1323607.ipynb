### Element-wise Operations

Handling missing values first

```python
sgpa_clean = pd.to_numeric(df['sgpa'], errors='coerce')
sgpa_clean = sgpa_clean.dropna()

# sgpa_clean.values + 1
```

#### Select Relevant Data

*Department-wise average SGPA*

```python
df_bar = df[['department', 'sgpa']]
```

Convert SGPA to numeric and remove missing values

```python
df_bar = df[['department', 'sgpa']].copy()

df_bar['sgpa'] = pd.to_numeric(df_bar['sgpa'], errors='coerce')
df_bar = df_bar.dropna()
```

Department-wise Average SGPA

```python
dept_avg_sgpa = df_bar.groupby('department')['sgpa'].mean()
avgsgpa = dept_avg_sgpa
dept_avg_sgpa
```

### Create the Bar Chart

```python
colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan', 'brown']
```

```python
plt.figure(figsize=(8,5))
plt.bar(dept_avg_sgpa.index, avgsgpa.values)
# plt.bar(dept_avg_sgpa['department'], dept_avg_sgpa['sgpa'], color=colors[:len(dept_avg_sgpa)])
plt.xlabel("Department")
plt.ylabel("Average SGPA")
plt.title("Department-wise Average SGPA")
plt.xticks(rotation=45)
plt.show()
```

Plot Using Column Names (NO ERROR)

```python
plt.figure(figsize=(8,5))
plt.bar(dept_avg_sgpa['department'], dept_avg_sgpa['sgpa'])
plt.xlabel("Department")
plt.ylabel("Average SGPA")
plt.title("Department-wise Average SGPA")
plt.xticks(rotation=45)
plt.show()
```
