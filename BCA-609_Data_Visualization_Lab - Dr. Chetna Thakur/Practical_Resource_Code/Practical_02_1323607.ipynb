### Program No. 02: Line Chart for Time-Based Data

Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Load the Student CSV Dataset

```python
file = "Uncleaned_Student_Dataset.csv"
df = pd.read_csv(file)
df.head()
```

Academic Year vs Attendance Percentage

```python
df_line = df[['academic_year', 'attendance_percentage']].copy()
```

Convert attendance to numeric

```python
df_line['attendance_percentage'] = pd.to_numeric(
    df_line['attendance_percentage'], errors='coerce'
)
```

Remove missing values

```python
df_line = df_line.dropna()
```

#### Group Data by Academic Year

average attendance per academic year

```python
yearly_attendance = (
    df_line
    .groupby('academic_year')['attendance_percentage']
    .mean()
    .reset_index()
)
yearly_attendance
```

```python
yearly_attendance = yearly_attendance.sort_values('academic_year')
```

#### Create the Line Chart

```python
plt.figure(figsize=(8,5))
plt.plot(
    yearly_attendance['academic_year'],
    yearly_attendance['attendance_percentage'],
    marker='o',
    color='green'
)

plt.xlabel("Academic Year")
plt.ylabel("Average Attendance (%)")
plt.title("Year-wise Average Student Attendance Trend")
plt.grid(True)
plt.show()
```
