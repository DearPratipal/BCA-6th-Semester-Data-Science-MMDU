Import dependecies library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Load the Dataset

```python
file = "Uncleaned_Student_Dataset.csv"
df = pd.read_csv(file)
df.head()
```

Performe the Operation

```python
df_rep = df[['department', 'sgpa']].copy()
```

```python
df_rep['sgpa'] = pd.to_numeric(df_rep['sgpa'], errors='coerce')
df_rep = df_rep.dropna()
```

Group Data (Department-wise Average SGPA)

```python
dept_avg = (
    df_rep
    .groupby('department')['sgpa']
    .mean()
    .reset_index()
)
dept_avg
```

#### Data Representation (Bar Chart)

Bar Chart – Clear Comparison

```python
plt.figure(figsize=(8,5))
plt.bar(dept_avg['department'], dept_avg['sgpa'], color='skyblue')
plt.xlabel("Department")
plt.ylabel("Average SGPA")
plt.title("Data Representation: Department-wise Average SGPA (Bar Chart)")
plt.xticks(rotation=45)
plt.show()
```

Pie Chart – Visual Presentation

```python
plt.figure(figsize=(6,6))
plt.pie(
    dept_avg['sgpa'],
    labels= dept_avg['department'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Data Presentation: Department-wise SGPA Distribution (Pie Chart)")
plt.show()
```
